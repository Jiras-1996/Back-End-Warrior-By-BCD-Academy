The Rick and Morty API is a REST(ish) and GraphQL API based on the television show Rick and Morty. You will have access to about hundreds of characters, images, locations and episodes. The Rick and Morty API is filled with canonical information as seen on the TV show.

## Step 02 : Import Data with API Requests

```python
# import library
import pandas as pd
import requests
import time

# Create list to contain data from API
name = []
status = []
species = []
type = []
gender = []
origin = []
location = []

for i in range(826):

    # Assign API url to variable
    url = f"https://rickandmortyapi.com/api/character/{i+1}"

    # HTTP Requests from url
    resp = requests.get(url)
    if resp.status_code == 200:
        json_data =  resp.json()
        name.append(json_data["name"])
        status.append(json_data["status"])
        species.append(json_data["species"])
        type.append(json_data["type"])
        gender.append(json_data["gender"])
        origin.append(json_data["origin"]["name"])
        location.append(json_data["location"]["name"])

    # Break requests a second to protect servers from overload
    #time.sleep(1)
```

## Step 03 : Create dataframe with Data from API and write to csv file

```python
# Create dataframe
rickandmorty_df = pd.DataFrame(data= {
    "name" : name,
    "status" : status,
    "species" : species,
    "type" : type,
    "gender" : gender,
    "origin" : origin,
    "location" : location
})

rickandmorty_df
```

```python
# Write data to csv file
rickandmorty_df.to_csv('drive/MyDrive/dataset/rickandmorty_df.csv', index=False)
```
